```python
import pandas as pd
import os

mainpath = r"C:\Users\Alexander\Documents\MLE_ProyectoFinal\data\raw"
filename = "Data_cov.csv"
fullpath = os.path.join(mainpath, filename)
```

```python
df = pd.read_csv(fullpath, encoding='latin1', sep=';')
df.head(10)
```

```python
df.Impactado = df.Impactado.replace({0: "No", 1: "Si"})
X = df.iloc[:, 3:16]
y = df.Impactado
```

```python
X_std = StandardScaler().fit_transform(X)
```

```python
train_x, test_x, train_y, test_y = train_test_split(X_std, y, test_size=0.2, random_state=42)
```

```python
best = SelectKBest(k=5)
X_train_new = best.fit_transform(train_x, train_y)
```

```python
with open('selectkbest_model.pkl', 'wb') as f:
    pickle.dump(best, f)
```
